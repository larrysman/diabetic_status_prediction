###### Model Deployment
 - Data Acquisition and Processing Strategy:

How we approach the deployment of a model and data is what this concept talks about. Will the model be sitting on streaming data or batchwise data. There are instances where data are not real time and company often gather the dataset at some point and push to the model and then recieves the predictions. This is batchwise approach of acquiring data and if this is the case, we do not need a model that will be live for making predictions.
On the otherhand, a model might be sitting on a streaming data and that means that the model needs to be exposed to the data in real time and make predicitons real time. Users can send in there queries into the model and the model makes predictions immediately and users get the response real time. With this in mind, there will be need to get a user interface where users get real time feedback on their usage of the model. Eg: In an insurance business, customers that want to process claims and access account, a model needs to be built on  a live transactional dataset of the insurance company.
The larger the size of a model, the higher the latency of making predictions - latency is the time it takes for a model to make predictions. This is a major consideration on a streaming data acquisition strategy.

 - Supported Process:

Which process is the model supporting. Is it supporting management decision making, business optimization, customers preference etc. If the model is supporting non transactional processes - for decision making, it is advisable to use the asynchronous data acquisition strategy which means batchwise data acquisition but if it is transactional processes - process that runs continously as part of the business, then the model will sit on streaming data and here the size of the model which contains weights and some artifacts should also be considered and must not be too big for effectiveness.
If an individual is using your model, you can use API but you need to be sure too because API calls fails.

 - Compute Requirement and Model Size:

Compute requirement is about the infrastructure your model is run. The bigger the model size, the more robust, the compute requirement should be. The computing power also reduces the latency of a model.
Model loading and offloading and inferencing time are affected by the computing resources.

 - Use Cases:

This speaks to what the model is suppose to produce and how critical is it to the process relying upon it, is expected to do.
In practise, if a model takes longer time to make predictions, a statistical equivalents of the model are created which can give a close inference and the process can revise its calculations. For transaction sensitive processes we dont do that.

#### <font color = 'grey'> Data Preprocessing Pipeline

```python
def preprocessor_training(data, target_col, scaler, min_max, norm):
    target = data[[target_col]]
    features = data.drop(columns=[target_col])
    num_data_transformed = scaler.fit_transform(features)
    num_data_transformed = min_max.fit_transform(num_data_transformed)
    num_data_transformed = norm.fit_transform(num_data_transformed)
    num_df = pd.DataFrame(num_data_transformed, columns=features.columns)
    target_df = target.reset_index(drop=True)
    preprocessed_df = pd.concat([num_df, target_df], axis=1)
    return preprocessed_df
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, Normalizer
from sklearn.model_selection import train_test_split
```

##### <font color = 'grey'> Correcting the data imbalance in the target variable

```python
def imbalance_corrector(data, target_col, strategy):
    target_df = data[[target_col]]
    feat_df = data.drop(columns=[target_col])
    feat_imb, target_imb = strategy.fit_resample(feat_df, target_df)
    return feat_imb, target_imb   
```

```python
from imblearn.combine import SMOTEENN
from imblearn.over_sampling import SMOTE
```

```python
smote_enn = SMOTEENN()
smote_enn
```

```python
data = preprocessed_df

target_col = 'Outcome'

strategy = smote_enn

feat_imb, target_imb = imbalance_corrector(data, target_col, strategy)
```

###### <font color = 'grey'> Predicition Function

###### The new preprocessing and prediction function

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, Normalizer
scaler
min_max
norm

import os
import joblib
model_path = model_path
os.chdir(model_path)
model = joblib.load('diabetes_model_df_v1_24-07-24 13-54-22.pkl')


def prep_prediction_function(Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction, Age):
    cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
    labels = ['No_Diabetis', 'Has_Diabetis']
    input = [Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction, Age]
    input_array = np.array(input).reshape(1, 8)
    num_data_transformed = scaler.transform(input_array)
    num_data_transformed = min_max.transform(num_data_transformed)
    num_data_transformed = norm.transform(num_data_transformed)
    num_df = pd.DataFrame(num_data_transformed, columns=cols)

    predictions = model.predict(num_df)
    
    if predictions[0] == 0:
        predicted_class = labels[0]
    else:
        predicted_class = labels[1]

    pred_df = pd.DataFrame({'predicted_outcome': [predictions[0]], 'patient_status': [predicted_class]}, columns=['predicted_outcome', 'patient_status'])

    return pred_df
```

```python
Pregnancies = 0.37160046
Glucose = 0.40412737
BloodPressure = 0.47455733
SkinThickness = 0.31777181
Insulin = 0.19830343
BMI =  0.45267701
DiabetesPedigreeFunction = 0.2018213
Age = 0.29217682

pred_df = prep_prediction_function(Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction, Age)
```
